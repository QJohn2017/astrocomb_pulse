```python
from scipy.io import loadmat
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = loadmat('smf_vs_edfa_data.mat')
shape = (len(np.unique(data['powers'])), len(np.unique(data['smfs'])))

smf = data['smfs'].reshape(*shape).T
power = data['powers'].reshape(*shape).T

fwhm = data['fwhms'].reshape(*shape).T
peaks = data['peaks'].reshape(*shape).T
```

```python
%matplotlib notebook

extent = [power.min(), power.max()+0.1, smf.min(), smf.max()+0.1]
plt.imshow(peaks, aspect='auto', origin='lower', extent=extent, cmap='inferno')
plt.xlabel('EDFA Power (W)')
plt.ylabel('SMF Compressor Length (m)')
plt.colorbar(label='Pulse Peak Power (pJ)')
plt.title('SMF 200m, HNLF 10m')
plt.savefig('edfa_vs_smf.png')
plt.show()
```

```python
%matplotlib notebook

plt.imshow(fwhm, aspect='auto', origin='lower', extent=extent, cmap='inferno')
plt.colorbar()
plt.show()
```
